```python
import pandas as pd
import numpy as np

# Load the data
data_FIFO = pd.read_excel('Flow Shop Assignment_Nguyen Xuan Binh_887799.xlsx', sheet_name='FIFO priority result', engine='openpyxl', skiprows=1)
data_EDD = pd.read_excel('Flow Shop Assignment_Nguyen Xuan Binh_887799.xlsx', sheet_name='EDD priority result', engine='openpyxl', skiprows=1)
```

```python
data_FIFO.head(10)
```

```python
group = 17

# Filter out pandas dataframe rows where column Group# is 17
data_FIFO = data_FIFO[data_FIFO['Group#'] == group]
data_FIFO.head(10)

data_EDD = data_EDD[data_EDD['Group#'] == group]
data_EDD.head(10)
```

```python
DD_FIFO = data_FIFO['DD'].tolist()
DD_EDD = data_EDD['DD'].tolist()

P_FIFO = data_FIFO.iloc[:, 2:6].values.tolist()
P_EDD = data_EDD.iloc[:, 2:6].values.tolist()
```

```python
# New lines for P
with open('FIFO priority_Nguyen Xuan Binh_887799.dat', 'w') as file:
    file.write(f"""I = 10; 				// Number of jobs\nK = 4; 				// Number of process stages\nM = 10000;				// Large number M\n""")
    file.write(f"DD = {DD_FIFO};\n")
    file.write("P = [\n")
    for row in P_FIFO:
        file.write(f"    {row},\n")
    file.write("];\n")

with open('EDD priority_Nguyen Xuan Binh_887799.dat', 'w') as file:
    file.write(f"""I = 10; 				// Number of jobs\nK = 4; 				// Number of process stages\nM = 10000;				// Large number M\n""")
    file.write(f"DD = {DD_EDD};\n")
    file.write("P = [\n")
    for row in P_EDD:
        file.write(f"    {row},\n")
    file.write("];\n")
```

```python
# No new lines for P
with open('FIFO priority_Nguyen Xuan Binh_887799.txt', 'w') as file:
    file.write(f"""I = 10; 				// Number of jobs\nK = 4; 				// Number of process stages\nM = 10000;				// Large number M\n""")
    file.write(f"DD = {DD_FIFO};\n")
    file.write(f"P = {P_FIFO};")

with open('EDD priority_Nguyen Xuan Binh_887799.txt', 'w') as file:
    file.write(f"""I = 10; 				// Number of jobs\nK = 4; 				// Number of process stages\nM = 10000;				// Large number M\n""")
    file.write(f"DD = {DD_EDD};\n")
    file.write(f"P = {P_EDD};")
```
